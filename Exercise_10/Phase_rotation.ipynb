Phase rotation

```python
from qiskit import QuantumCircuit
from qiskit.visualization import plot_bloch_multivector
from qiskit.quantum_info import Statevector
import numpy as np
```

```python
qc = QuantumCircuit(1)

initial_state = Statevector.from_instruction(qc)
print("Initial state:")
display(plot_bloch_multivector(initial_state))
```

```python
qc.h(0)

state_after_h = Statevector.from_instruction(qc)
print("State after Hadamard Gate:")
display(plot_bloch_multivector(state_after_h))
```

```python
# Apply state rotation
#qc.p(np.pi , 0) #180 degrees
#qc.p(np.pi/4 , 0) #45 degrees
#qc.p(np.pi/2 , 0) #90 degrees
qc.p(3 * np.pi / 2, 0) # 270 degrees 

final_state = Statevector.from_instruction(qc)

print("State after Phase Rotation:")
display(plot_bloch_multivector(final_state))
```

```python
# Funcin para calcular y mostrar las coordenadas de Bloch
def print_bloch_coordinates(state, label):
    x = 2 * np.real(state[0] * np.conj(state[1]))
    y = 2 * np.imag(state[0] * np.conj(state[1]))
    z = np.abs(state[0])**2 - np.abs(state[1])**2
    display(f"Bloch coordnates ({label}): x = {x:.3f}, y = {y:.3f}, z = {z:.3f}")
```

```python
print_bloch_coordinates(initial_state, "Initial")
print_bloch_coordinates(state_after_h, "After Hadamard")
print_bloch_coordinates(final_state, "Final")
```
